```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.animation as animation
import time
```

```python
df = pd.read_csv("outfile.txt", sep=",", nrows = 100*10000, usecols = [0,1], header=None)
A = df.astype(float).to_numpy();
```

```python
x = A[:10000,0];
y = A[:10000,1];
fig, ax = plt.subplots()
scat = ax.scatter(x, y, c='red', s=1, marker = ',')  # Initial scatter plot
```

```python
def update(frame):
    # Get the next 100 rows (frame is the index)
    start = frame * 10000
    end = (frame + 1) * 10000
    scat.set_offsets(A[start:end])
    return scat,
```

```python
#ani = animation.FuncAnimation(fig, update, frames=range(A.shape[0] // 10000),
 #                             interval=200, blit=True)
plt.ion()
for i in range(100):
    start = i * 10000
    end = (i + 1) * 10000
    scat.set_offsets(A[start:end])
    display(fig)
plt.show()
```
